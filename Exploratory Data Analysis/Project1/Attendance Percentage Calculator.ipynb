```python
import pandas as pd
```

```python
# Reading the Attendance data from given Excel file.

df1 = pd.read_excel("Attendance.xlsx", sheet_name = "PE")
```

```python
df = df1.copy()
df.head()
```

```python
# Cleaning The Dataset

df.drop(labels = [ 0, 1, 2, 34, 35, 36, 37, 38, 39],inplace = True)

df.reset_index(inplace = True)

df.drop(labels = ["index", "Unnamed: 3", "Unnamed: 52","Unnamed: 53","Unnamed: 54", "Unnamed: 55"],inplace = True, axis = 1)
```

```python
# Giving New column names

x = [f"Day {i}"  for i in range(1, 49 )]

y = ["Roll No", "Name", "Attendance %"]

df.columns = y + x

df.head()
```

```python
df2 = df.drop(axis = 1, labels = ["Roll No", "Name", "Attendance %"])
df2
```

```python
def refresher(x):
    mean = 0
    for i in x:
        if i == "P":
            mean = mean + 1
    return (mean / len(x)) * 100
 
x = df2.apply(refresher, result_type = 'reduce', axis = 1)
df["New Attendance %"] = x
```
