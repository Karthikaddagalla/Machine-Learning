```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)
```

```python
df = pd.DataFrame()

df['X'] = X.reshape(100)
df['y'] = y

df
```

```python
df["pred"] = df["y"].mean()
df["Residual"] = df["y"] - df["pred"]

df
```

```python
# here we asking the model to predict amount of error it is doing
for i in range(0, 100):
    model = DecisionTreeRegressor()
    model.fit(df[["X"]], df[["Residual"]])
    df["pred"] = df["pred"] + 0.1*model.predict(df[["X"]])
    
    df["Residual"] = df["y"] -  df["pred"]    
    
    
```
